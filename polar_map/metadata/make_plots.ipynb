```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
from matplotlib import dates as mdates
```

```python
df = pd.read_csv('metadata_hmi_nonpolar.csv', skipinitialspace=True).dropna()
files = df['filename'].to_numpy()
#dates = [datetime.fromisoformat(file.split('_')[3]) for file in files]
dates = [datetime.fromisoformat(file.split('.')[2][:-4]) for file in files]

pxbeg1 = df['PXBEG1'].to_numpy()
pxbeg2 = df['PXBEG2'].to_numpy()
crpix1 = df['CRPIX1'].to_numpy()
crpix2 = df['CRPIX2'].to_numpy()
rsun_arc = df['RSUN_ARC'].to_numpy()
cdelt1 = df['CDELT1'].to_numpy()
dsun_au = df['DSUN_AU'].to_numpy()
crota = df['CROTA'].to_numpy()
crln_obs = df['CRLN_OBS'].to_numpy()
hgln_obs = df['HGLN_OBS'].to_numpy()
crlt_obs = df['CRLT_OBS'].to_numpy()
obs_vr = df['OBS_VR'].to_numpy() / 1000
obs_vw = df['OBS_VW'].to_numpy() / 1000
obs_vn = df['OBS_VN'].to_numpy() / 1000
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(dates, obs_vr, label='OBS VR')
ax.plot(dates, obs_vw, label='OBS VW')
ax.plot(dates, obs_vn, label='OBS VN')
ax.set_ylabel('S/C velocity, km/s')

locator = mdates.AutoDateLocator()
formatter = mdates.ConciseDateFormatter(locator)
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)

ax.legend(loc='center left')

ax_ = ax.twinx()
ax_.plot(dates, dsun_au, color='tab:red')
ax_.set_ylabel('Distance, AU', color='tab:red')


plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(dates, crpix1 + pxbeg1 - 1, label='CRPIX1')
ax.plot(dates, crpix2 + pxbeg2 - 1, label='CRPIX2')
ax.set_ylabel('Disk center, pix')

locator = mdates.AutoDateLocator()
formatter = mdates.ConciseDateFormatter(locator)
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)

ax.legend(loc='center left')

ax_ = ax.twinx()
ax_.plot(dates, rsun_arc / cdelt1, color='tab:green')
ax_.set_ylabel('Disk radius, pix', color='tab:green')

#plt.grid(True)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(dates, crln_obs, label='Carrington')
ax.plot(dates, hgln_obs % 360, label='Stonyhurst')
ax.set_ylabel('Heliographic longitude, degrees')

locator = mdates.AutoDateLocator()
formatter = mdates.ConciseDateFormatter(locator)
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)

ax.legend(loc='lower left')

ax_ = ax.twinx()
ax_.plot(dates, crlt_obs, color='tab:green')
ax_.set_ylabel('Heliographic latitude, degrees', color='tab:green')


plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(dates, crota)
ax.set_ylabel('Roll angle, degrees')

locator = mdates.AutoDateLocator()
formatter = mdates.ConciseDateFormatter(locator)
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)
plt.tight_layout()
```
