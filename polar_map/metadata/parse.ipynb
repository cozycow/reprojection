```python
from astropy.io import fits
import glob
```

```python
files = sorted(glob.glob('/home/ulyanov/data/solo/phi/2024_2025/*'))#files = sorted(glob.glob('/home/ulyanov/data/sdo/hmi/nonpolar/*'))
print(len(files))
print(files[0], files[-1])
```

```python
output_file = 'metadata_phi.csv'

with open(output_file, 'w') as f:
    f.write('DATE, DID, DSUN_AU, CAR_ROT, CRLN_OBS, HGLN_OBS, CRLT_OBS, OBS_VR, OBS_VW, OBS_VN\n')

for file in files[:]:
    with fits.open(file) as hdul:
        header = hdul[0].header.copy()

    file_ = file.split('/')[-1]

    date = file_.split('_')[3]
    did = file_.split('_')[-1].split('.')[0]


    dsun_au = header['DSUN_AU'] if 'DSUN_AU' in header else header['DSUN_OBS'] / header['DSUN_REF']
    car_rot = header['CAR_ROT']
    crln_obs = header['CRLN_OBS']
    hgln_obs = header['HGLN_OBS'] if 'HGLN_OBS' in header else 0
    crlt_obs = header['CRLT_OBS']
    obs_vr = header['OBS_VR']
    obs_vw = header['OBS_VW']
    obs_vn = header['OBS_VN']

    with open(output_file, 'a') as f:
        f.write(f'{date}, {did}, {dsun_au}, {car_rot}, {crln_obs}, {hgln_obs}, {crlt_obs}, {obs_vr}, {obs_vw}, {obs_vn}\n')

```
