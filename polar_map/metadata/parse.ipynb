```python
from astropy.io import fits
import glob
```

```python
#files = sorted(glob.glob('/home/ulyanov/data/solo/phi/nonpolar/*'))
files = sorted(glob.glob('/home/ulyanov/data/sdo/hmi/nonpolar/*'))
print(len(files))
print(files[0], files[-1])
```

```python
output_file = 'metadata_hmi_nonpolar.csv'

with open(output_file, 'w') as f:
    f.write('filename, PXBEG1, PXBEG2, CRPIX1, CRPIX2, RSUN_ARC, CDELT1, DSUN_AU, CROTA, CRLN_OBS, HGLN_OBS, CRLT_OBS, OBS_VR, OBS_VW, OBS_VN\n')

for file in files:
    with fits.open(file) as hdul:
        header = hdul[1].header.copy()

    file_ = file.split('/')[-1]

    pxbeg1 = header['PXBEG1'] if 'PXBEG1' in header else 1
    pxbeg2 = header['PXBEG2'] if 'PXBEG2' in header else 1
    crpix1 = header['CRPIX1']
    crpix2 = header['CRPIX2']
    rsun_arc = header['RSUN_ARC'] if 'RSUN_ARC' in header else header['RSUN_OBS']
    cdelt1 = header['CDELT1']
    dsun_au = header['DSUN_AU'] if 'DSUN_AU' in header else header['DSUN_OBS'] / header['DSUN_REF']
    crota = header['CROTA'] if 'CROTA' in header else header['CROTA2']
    crln_obs = header['CRLN_OBS']
    hgln_obs = header['HGLN_OBS'] if 'HGLN_OBS' in header else 0
    crlt_obs = header['CRLT_OBS']
    obs_vr = header['OBS_VR']
    obs_vw = header['OBS_VW']
    obs_vn = header['OBS_VN']

    with open(output_file, 'a') as f:
        f.write(f'{file_}, {pxbeg1}, {pxbeg2}, {crpix1}, {crpix2}, {rsun_arc}, {cdelt1}, {dsun_au}, '
                f'{crota}, {crln_obs}, {hgln_obs}, {crlt_obs}, {obs_vr}, {obs_vw}, {obs_vn}\n')

```
