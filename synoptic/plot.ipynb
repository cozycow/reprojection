```python
files = sorted(glob.glob('/home/ulyanov/data/sdo/hmi/synoptic_maps/*'))

with fits.open(files[0]) as hdul:
    header = hdul[0].header.copy()
    data = hdul[0].data.copy()
```

```python
lat = np.arcsin(np.arange(-1, 1 + 1 / 719.5 / 2, 1 / 719.5)) * 180 / np.pi

flux_density = np.nanmean(data, axis=1)
flux_density_ = np.nanmean(mean, axis=1)

plt.figure(figsize=(10,8))
plt.plot(lat, flux_density)
plt.plot(lat, flux_density_)

plt.xlim(-90,90)
plt.ylim(-8,8)

plt.grid(True)
plt.tight_layout()
```

```python
R = 695e8

t = np.where(lat > 70)
np.trapezoid(flux_density_[t] * np.sin(lat[t] * np.pi / 180), lat[t]) * R ** 2 * 2 * np.pi
```
