```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import glob
from transforms import *
from reprojection import *
```

```python
PHI_TOP = '/home/ulyanov/data/solo/phi/'
files = sorted(glob.glob(PHI_TOP + '/polar/*'))
```

```python
file1 = files[100]
file2 = files[120]

with fits.open(file1) as hdul:
    header1 = hdul[0].header
    data1 = hdul[0].data

with fits.open(file2) as hdul:
    header2 = hdul[0].header
    data2 = hdul[0].data
```

```python
view = View.from_header(header1)
transform = view.to_spherical()

grid = np.mgrid[-90:91:1,-180:180:1]
grid, _ = transform(grid)
grid = np.array(grid)

meridians = grid[:,:,::15]
parallels = np.transpose(grid[:,::15,:], (0,2,1))

plt.figure(figsize=(10,10))
plt.imshow(data1, origin='lower', cmap='seismic', vmin=-40, vmax=40)
plt.plot(meridians[1], meridians[0], color='black', ls='--', lw=1, alpha=0.5)
plt.plot(parallels[1], parallels[0], color='black', ls='--', lw=1, alpha=0.5)
plt.tight_layout()
```

```python
view = View.from_header(header2)
transform = view.to_spherical()

grid = np.mgrid[-90:91:1,-180:180:1]
grid, _ = transform(grid)
grid = np.array(grid)

meridians = grid[:,:,::15]
parallels = np.transpose(grid[:,::15,:], (0,2,1))

plt.figure(figsize=(10,10))
plt.imshow(data2, origin='lower', cmap='seismic', vmin=-40, vmax=40)
plt.plot(meridians[1], meridians[0], color='black', ls='--', lw=1, alpha=0.5)
plt.plot(parallels[1], parallels[0], color='black', ls='--', lw=1, alpha=0.5)
plt.tight_layout()
```

```python
data2_ = reproject(data2, header2, header1, correct_mu=True, mu_thr=0.)
```
