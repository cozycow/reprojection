```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import glob
from utils import *
from calc_fluxes import *
```

```python
files = sorted(glob.glob('/home/ulyanov/data/solo/phi/2024_2025/*'))
```

```python
Q = []
S = []


for yy in [2024, 2025]:
    for mm in range(1,13):
        print(yy, mm)
        files_ = []
        for file in files:
            date = int(file.split('/')[-1].split('_')[3][:8])

            if (date > yy * 10000 + mm * 100) and (date < yy * 10000 + mm * 100 + 31):
                files_.append(file)

        lat, flux_density, variance = calc_fluxes(files_)

        Q += [flux_density]
        S += [variance]

Q = np.array(Q)
S = np.array(S)
```

```python
plt.figure(figsize=(10,8))
plt.imshow(Q.T, 'bwr', aspect='auto', origin='lower', vmin=-10, vmax=10)
plt.tight_layout()
```

```python
np.savez('fluxes_monthly.npz', lat=lat, flux_density=Q, variance=S)
```
