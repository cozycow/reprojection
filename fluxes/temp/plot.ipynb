```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
import sunpy.visualization.colormaps as cm
from astropy.io import fits
import glob
```

```python
s = np.load('fluxes_202411fdt.npz')
flux_density_1 = s['flux_density']
variance_1 = s['variance']
lat_1 = s['lat'].astype(float)

dlat_1 = lat_1[1] - lat_1[0]
lat_1 += dlat_1 / 2
```

```python
s = np.load('fluxes_202411hmi.npz')
flux_density_2 = s['flux_density']
variance_2 = s['variance']
lat_2 = s['lat'].astype(float)

dlat_2 = lat_2[1] - lat_2[0]
lat_2 += dlat_2 / 2
```

```python
fig = plt.figure(figsize=(10,8))
plt.errorbar(lat_1, flux_density_1, yerr=np.sqrt(variance_1), label='FDT (2024-11)')
plt.errorbar(lat_2, flux_density_2, yerr=np.sqrt(variance_2), label='HMI (2024-11)')

plt.xlabel('Latitude, degrees')
plt.ylabel(r'Magnetic flux density, $Mx/cm^2$')

plt.legend()

plt.xlim(-90,90)
plt.ylim(-15,15)
plt.grid(True)
plt.tight_layout()
```

```python
R = 695e8

t = np.where(np.all([lat_1 > 40, lat_1 < 60], axis=0))
#t = np.where(lat_1 > 60)
F_fdt = np.trapezoid(np.abs(flux_density_1[t]) * np.sin(lat_1[t] * np.pi / 180), lat_1[t]) * R ** 2 * 2 * np.pi

t = np.where(np.all([lat_2 > 40, lat_2 < 60], axis=0))
#t = np.where(lat_2 > 66)
F_hmi = np.trapezoid(np.abs(flux_density_2[t]) * np.sin(lat_2[t] * np.pi / 180), lat_2[t]) * R ** 2 * 2 * np.pi
```
