```python
s = np.load('data/2302fdt.npz')
data_fdt = s['mean']
var_fdt = s['variance']
lon = s['lon']
lat = s['lat']
```

```python
s = np.load('data/2302hmi.npz')
data_hmi = s['mean']
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax.imshow(data_fdt, aspect='auto', origin='lower', cmap='hmimag',
           extent=(0, 360, -1, 1), vmin=-1000, vmax=1000)
ax.set_yticks(np.sin(np.arange(-90, 91, 15) * np.pi / 180), np.arange(-90, 91, 15))
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax.imshow(np.sqrt(var_fdt.clip(0)), aspect='auto', origin='lower',
           extent=(0, 360, -1, 1), vmin=0, vmax=10)
ax.set_yticks(np.sin(np.arange(-90, 91, 15) * np.pi / 180), np.arange(-90, 91, 15))
plt.tight_layout()
```

```python
flux_hmi = np.nanmean(data_hmi[:,:600], axis=-1)
flux_fdt = np.nanmean(data_fdt[:,:600], axis=-1)

dsine = 1 / 359.5
lat = np.arange(-1,1 + dsine / 2, dsine)
lat = np.arcsin(lat.clip(-1,1)) * 180 / np.pi

plt.figure(figsize=(10,8))
plt.plot(lat, flux_fdt, label='FDT')
plt.plot(lat, flux_hmi, label='HMI')

plt.xlim(0,90)
plt.ylim(-10,10)

plt.title('CR 2302, North pole')
plt.xlabel('Longitude, degrees')
plt.ylabel(r'Flux density, Mx / cm$^2$')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
