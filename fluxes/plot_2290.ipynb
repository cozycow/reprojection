```python
fig, ax = plt.subplots(figsize=(14,8))
ax.imshow(data_fdt, aspect='auto', origin='lower', cmap='hmimag',
           extent=(0, 360, -1, 1), vmin=-1000, vmax=1000)
ax.set_yticks(np.sin(np.arange(-90, 91, 15) * np.pi / 180), np.arange(-90, 91, 15))
plt.tight_layout()
```

```python
flux_hmi = np.nanmean(data_hmi, axis=-1)
flux_fdt = np.nanmean(data_fdt, axis=-1)

dsine = 1 / 359.5
lat = np.arange(-1,1 + dsine / 2, dsine)
lat = np.arcsin(lat.clip(-1,1)) * 180 / np.pi

plt.figure(figsize=(10,8))
plt.plot(lat, flux_fdt, label='FDT')
plt.plot(lat, flux_hmi, label='HMI')

plt.xlim(-90,90)
plt.ylim(-10,10)

plt.title('CR 2290/2291, North pole')
plt.xlabel('Longitude, degrees')
plt.ylabel(r'Flux density, Mx / cm$^2$')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
flux_hmi = np.nanmean(np.abs(data_hmi), axis=-1)
flux_fdt = np.nanmean(np.abs(data_fdt), axis=-1)

dsine = 1 / 359.5
lat = np.arange(-1,1 + dsine / 2, dsine)
lat = np.arcsin(lat.clip(-1,1)) * 180 / np.pi

plt.figure(figsize=(10,8))
plt.plot(lat, flux_fdt, label='FDT')
plt.plot(lat, flux_hmi, label='HMI')

plt.xlim(-90,90)
plt.ylim(0,50)

plt.title('CR 2290/2291, North pole')
plt.xlabel('Longitude, degrees')
plt.ylabel(r'Flux density, Mx / cm$^2$')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
q_hmi = np.nanmean(data_hmi, axis=-1) / np.nanmean(np.abs(data_hmi), axis=-1)
q_fdt = np.nanmean(data_fdt, axis=-1) / np.nanmean(np.abs(data_fdt), axis=-1)

dsine = 1 / 359.5
lat = np.arange(-1,1 + dsine / 2, dsine)
lat = np.arcsin(lat.clip(-1,1)) * 180 / np.pi

plt.figure(figsize=(10,8))
plt.plot(lat, q_fdt, label='FDT')
plt.plot(lat, q_hmi, label='HMI')

plt.xlim(-90,90)
plt.ylim(-1.2,1.2)

plt.title('CR 2290/2291, North pole')
plt.xlabel('Longitude, degrees')
plt.ylabel(r'Flux density, Mx / cm$^2$')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
for lat0 in np.arange(0,75,15):
    t = np.where(np.all([lat > lat0, lat < lat0 + 15], axis=0))

    temp1 = np.nanmean(np.abs(data_fdt), axis=-1)
    temp2 = np.nanmean(np.abs(data_hmi), axis=-1)

    F1, F2 = (np.trapezoid(temp1[t] * np.cos(lat[t] * np.pi / 180), x=lat[t]),
              np.trapezoid(temp2[t] * np.cos(lat[t] * np.pi / 180), x=lat[t]))

    print(f'{lat0}-{lat0+15}:', F1, F2, F1 / F2)
```
